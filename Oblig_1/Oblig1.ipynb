1) Det er $2^N$ antall microstates fordi hver partikkel har 2 spinmuligheter og det er $N$ partikkler.

2) $S =\sum_i^Ns_i$

```python
#3)

import numpy as np
import random
import matplotlib.pyplot as plt

M = int(1e4); N = 60
S = np.zeros(M).astype(int)

for j in range(0,M):
    for i in range(0,N):
        S[j] += random.choice([-1,1])

#setter opp gaussisk sannsynlighetsfunksjon
mean_S = np.sum(S)/M
sum = 0
for i in range(0,M):
    sum  += (S[i]-mean_S)**2

sigma = np.sqrt(1/(M-1)*sum)

h = np.linspace(-25,25,M)
P = 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(h-mean_S)**2/(2*sigma**2))

plt.plot(h,P,linewidth=2)
plt.hist(S, bins = 30,edgecolor='black',density=True)
plt.xlabel('S',FontSize='25')
plt.ylabel('Intensitet',FontSize='25')
plt.show()
```

Ser at totalspinnet $S$ er omtrent gaussisk for $M=10^4$. Siden det er tilfeldig hvilken verdi $s_i\in${$-1,1$} får vil vi forvente en normalfordelt kurve om $S=0$ som blir gaussisk når $M\rightarrow\infty$.
